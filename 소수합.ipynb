```python
def guguVertical():
    for n in range(1, 10):  # 단 수가 밖에 존재.
        for m in range(2, 10):
            print(f'{n} x {m} = {n * m}')  # f스트링 이용

guguVertical()
```

```python
def guguHorizontal():
    for m in range(2, 10):
        for n in range(1, 10):  # 단 수가 안에 들어왔다.
            print(f'{n} x {m} = {n * m}', end='\t')
        print()   
        
       
```
